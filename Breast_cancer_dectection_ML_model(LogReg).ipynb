```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression 
import sklearn.datasets as skd
```

```python
data = pd.read_csv(r'data.csv')
data
```

```python
breast_cancer_sk = skd.load_breast_cancer()
```

```python
print(breast_cancer_sk)
```

```python
df= pd.DataFrame(breast_cancer_sk.data,columns=breast_cancer_sk.feature_names)
df
```

```python
df["label"]=breast_cancer_sk.target
```

```python
df
```

```python
df.describe()
```

```python
df.groupby(df.label).mean()
```

```python
df.head()
```

```python
df.shape
```

```python
X=df.drop(columns='label')
```

```python
X
```

```python
Y = df['label']
```

```python
Y
```

```python
x_train, x_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
x_test
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred = model.predict(x_train)
x_test_pred= model.predict(x_test)
```

```python
acc_test = accuracy_score(y_test,x_test_pred)
acc_train = accuracy_score(y_train,x_train_pred,)
print('Acc on test =', acc_test)
print('Acc on train =', acc_train)
```
